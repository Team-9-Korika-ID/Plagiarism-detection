```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize

# Dataframe
data_lite = {'Nama': ['Andi', 'Budi', 'Anton'],
             'Jawaban': ['Revolusi Industri 4.0 merupakan periode transformasi industri yang ditandai oleh integrasi teknologi digital ke dalam proses produksi.',
                         'Revolusi Industri 4.0 merupakan periode transformasi industri yang ditandai oleh integrasi teknologi digital ke dalam proses produksi.',
                         'Revolusi Industri 4.0 merupakan upaya transformasi menuju perbaikan dengan mengintegrasikan dunia online dan lini produksi di industri']}
df = pd.DataFrame(data_lite)

# tf-idf
tfidf_vectorizer = CountVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Jawaban'])
tfidf_matrix_normalized = normalize(tfidf_matrix)

# Similarity Matrix
similarity_matrix = cosine_similarity(tfidf_matrix_normalized, tfidf_matrix_normalized)

# Menampilkan hasil similarity
hasil = pd.DataFrame(similarity_matrix, columns=df['Nama'], index=df['Nama'])

# Menerapkan logika dan menambahkan warna
def highlight_cells(val):
    color = 'red' if val == 'Mencontek' else 'black'
    return 'color: %s' % color

# Menambahkan penanda berdasarkan nilai similarity
for index, row in hasil.iterrows():
    for column in hasil.columns:
        if index != column:
            if row[column] >= 0.9:
                hasil.at[index, column] = 'Mencontek'
            elif row[column] <= 0.6:
                hasil.at[index, column] = 'Original'

# Menampilkan dataframe dengan warna
hasil.style.applymap(highlight_cells)
```

```python
from transformers import AutoTokenizer, AutoModel
from sklearn.metrics.pairwise import cosine_similarity

# Load pre-trained transformer model and tokenizer
model_name = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Encode sentences into tensors
sentence_1 = "The cat sat on the mat."
sentence_2 = "The dog lay on the rug."
encoded_1 = tokenizer(sentence_1, return_tensors='pt')
encoded_2 = tokenizer(sentence_2, return_tensors='pt')

# Get model outputs
with torch.no_grad():
    model_output_1 = model(**encoded_1)
    model_output_2 = model(**encoded_2)

# Extract sentence embeddings
sentence_embedding_1 = model_output_1.pooler_output
sentence_embedding_2 = model_output_2.pooler_output

# Calculate cosine similarity
similarity = cosine_similarity([sentence_embedding_1.detach().numpy()], [sentence_embedding_2.detach().numpy()])[0][0]

# Print the result
print(f"Similarity between sentences: {similarity}")
```

## Resume screening

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Persyaratan pekerjaan
requirement = """
Persyaratan Pekerjaan:
- Minimal 2 tahun pengalaman di bidang yang sesuai
- Usia minimal 25 tahun
- Lulusan Teknik Informatika atau bidang terkait
"""

# Resume kandidat
resume_kandidat = [
    "John Doe memiliki pengalaman kerja selama 5 tahun di bidang teknologi informasi.",
    "Jane Smith adalah lulusan Teknik Informatika dan memiliki pengalaman kerja selama 1 tahun.",
    "Ahmad Yani memiliki pengalaman kerja selama 1 tahun di bidang teknologi informasi.",
]

def calculate_similarity(requirement, resume):
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform([requirement] + resume)
    similarity_matrix = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:])
    return similarity_matrix

def main():
    similarity_scores = calculate_similarity(requirement, resume_kandidat)
    max_similarity_index = similarity_scores.argmax()
    max_similarity_score = similarity_scores[0][max_similarity_index]
    print(f"Kandidat terpilih: Kandidat {max_similarity_index + 1} dengan skor kesesuaian {max_similarity_score}")

if __name__ == "__main__":
    main()
```
